# Multivariate Gaussian with full covariance

- TensorFlow를 사용하여 다변량 가우시안 분포를 지정하는 방법을 배웁니다.  

```python
import seaborn as sns
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow_probability as tfp
tfd = tfp.distributions
```

`tfd.MultivariateNormalDiag` 클래스를 사용하면 대각 공분산 행렬 $\Sigma$를 사용하여 다변량 가우시안을 지정할 수 있습니다.

분산이 각 성분에 대해 동일한 경우, 즉 $\Sigma=\sigma^2I$인 경우, 이것은 구형 또는 등방성(isotropic) 가우시안으로 알려져 있습니다. 이 이름은 확률 밀도 함수의 구형(또는 원형) 등고선에서 비롯된 것입니다. 2 차원 사례에 대한 아래 그림에서 볼 수 있습니다.

## tfd.Normal

- The Normal distribution with location loc and scale parameters.

```python
# Define a batch of two scalar valued Normals.
dist = tfd.Normal(loc=[1, 2.], scale=[11, 22.])

dist.mean(), dist.stddev()
```

```python
# Plot the approximate density contours of a 2d spherical Gaussian
N = 1000
x = dist.sample(N)
x1 = x[:, 0]
x2 = x[:, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0)
```

## tfd.Independent
- Independent distribution from batch of distributions  

- reinterpreted_batch_ndims:   
    Scalar, integer number of rightmost batch dims which will be regarded as event dims. When `None` all but the first batch axis (batch axis 0) will be transferred to event dimensions (analogous to `tf.layers.flatten`).

```python
# Make independent distribution from a 2-batch Normal.
ind = tfd.Independent(
    distribution=tfd.Normal(loc=[1, 2.], scale=[11, 22.]),
    reinterpreted_batch_ndims=1)

ind.mean(), ind.stddev()
```

```python
# Plot the approximate density contours of a 2d spherical Gaussian
N = 1000
x = ind.sample(N)
print(x.shape)

x1 = x[:, 0]
x2 = x[:, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0)
```

```python
# Make independent distribution from a 2-batch bivariate Normal.
ind = tfd.Independent(
    distribution=tfd.MultivariateNormalDiag(
        loc=[[-1., 1], [1, -1]],
        scale_identity_multiplier=[1., 0.5]),
    reinterpreted_batch_ndims=1)
ind
```

```python
# Plot the approximate density contours of a 2d spherical Gaussian
N = 1000
x = ind.sample(N)
print(x.shape)

x1 = x[:, 0, 0]
x2 = x[:, 0, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0)
```

## tfd.MultivariateNormalDiag

- The multivariate normal distribution on $R^k$.

```python
# Define multivariate Normal distribution.
mvn = tfd.MultivariateNormalDiag(
    loc=[1., 2.],
    scale_diag=[11., 22.])

mvn.mean(), mvn.stddev()
```

```python
# Probability density/mass function.
mvn.prob([-1., 0])
```

```python
# Plot the approximate density contours of a 2d spherical Gaussian
N = 1000
x = mvn.sample(N)
print(x.shape)

x1 = x[:, 0]
x2 = x[:, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0)
```

- 대각 공분산 행렬은 랜덤 벡터의 성분이 독립적이 되도록합니다.

# Full covariance with MultivariateNormalFullTriL

- tfd.MultivariateNormalTriL 분포를 사용하여 TensorFlow에서 전체 공분산 가우시안 분포를 정의할 수 있습니다.

- 수학적으로 다변량 가우시안의 매개 변수는 평균 $\mu$와 공분산 행렬 $\Sigma$입니다. 따라서 tfd.MultivariateNormalTriL 생성자에는 두 개의 인수가 필요합니다.
    - loc : $\mu$ 에 해당하는 float Tensor
    - scale_tril : 하부 삼각 행렬 $L$ ($LL^T=\Sigma$)  
    
- d 차원 random variable 의 경우 lower-triangle matrix $L$은 다음과 같습니다.
<img src="lower-triangle.png" width=200>

- 다음은 비 대각선 공분산(non-diagonal covariance)으로 2 차원 가우시안을 생성하는 예입니다.

```python
# Set the mean and covariance parameters
mu = [0., 0.]  # mean
scale_tril = [[1.,  0.], [0.6, 0.8]]

# covariance matrix
sigma = tf.matmul(tf.constant(scale_tril), tf.transpose(tf.constant(scale_tril)))  
print(sigma)
```

```python
tfp.layers.MultivariateNormalTriL
```

```python
# Create the 2D Gaussian with full covariance
nonspherical_2d_gaussian = tfd.MultivariateNormalTriL(loc = mu, scale_tril = scale_tril)
nonspherical_2d_gaussian
```

```python
# Check the Distribution mean
print(nonspherical_2d_gaussian.mean())

# Check the Distribution covariance
print(nonspherical_2d_gaussian.covariance())
```

```python
# Plot its approximate density contours
x = nonspherical_2d_gaussian.sample(N)
x1 = x[:, 0]
x2 = x[:, 1]
sns.jointplot(x=x1, y=x2, kind = 'kde', space = 0, color = 'r')
```

- 보시다시피, 대략적인 밀도 윤곽은 이제 원형이 아니라 타원형입니다. 이는 가우시안의 구성 요소가 상관 관계가 있기 때문입니다.  


- 또한 주변 분포(그림의 측면에 표시됨)는 둘 다 단변량 가우시안 분포입니다.
